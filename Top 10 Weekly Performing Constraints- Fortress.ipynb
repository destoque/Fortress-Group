```python
#dropping the unecessary columns

rtbinds = rtbinds.drop(['mon_fac','cont_fac','cons_type','nercid','state', 'hour'], axis=1)
```

```python
#new dataframe with relevant columns

rtbinds
```

```python
#removing the "0:00" string so I can convert the "pricedate" column to datetime

rtbinds['pricedate'] = rtbinds['pricedate'].str.replace('0:00', '')
```

```python
#Converting to datetime
rtbinds['pricedate'] =  pd.to_datetime(rtbinds['pricedate'])
```

```python
#creating columns for year and week
rtbinds['year'] = rtbinds['pricedate'].apply(lambda x: x.year)
rtbinds['week'] = rtbinds['pricedate'].apply(lambda x: x.isocalendar()[1])
```

```python
#Creating a new dataframe to grab relevant values
rtbinds2 = rtbinds.groupby(['year', 'week', 'cons_name'], as_index=False).aggregate(np.sum)
```

```python
#creating a function to present the top 10 values (by sum) of each week

def udf(rtbinds):
    return rtbinds.sort_values('shadow').head(10)

#running function in new df
rtbinds2.groupby(['year', 'week'], as_index=False).apply(udf)
```
